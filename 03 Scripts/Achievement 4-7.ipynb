## Price Label

```python
def price_label(row):

    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

```python
df=df.reset_index(drop=True) #Kept getting same warning as above when using .loc, had to use this command to stop warnings.
```

```python
df.head(10)
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

## Busiest Day

```python
result=[]

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

## Task

```python
#This label should correspond with the two busiest days of the week as opposed to the single busiest day.
#they’d also like to know the two slowest days
result=[]

for value in df_ords_prods_merged["orders_day_of_week"]:
    if value == 0:
        result.append("Busiest day 1")
    elif value == 1:
        result.append("Busiest day 2")
    elif value == 3:
        result.append("Slowest day 1")
    elif value == 4:
        result.append("Slowest day 2")
    else:
        result.append("Regularly busy")
```

```python
df_ords_prods_merged['busiest days']=result
```

```python
df_ords_prods_merged['busiest days'].value_counts(dropna=False)
```
